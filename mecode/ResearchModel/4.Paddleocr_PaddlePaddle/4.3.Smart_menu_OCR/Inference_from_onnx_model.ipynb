```python
# script: to extract text from output.txt

import csv
import json

# Function to process a line and return filename and concatenated transcriptions
def process_line(line):
    # Splitting the line into file name and JSON data
    file_name, json_data = line.split('\t')

    # Parse the JSON data
    parsed_data = json.loads(json_data)

    # Concatenate the transcriptions
    transcriptions = ' '.join([item['transcription'] for item in parsed_data])

    return file_name, transcriptions

# File path of your .txt file
input_file_path = '/content/output_imgv1/system_results.txt'
# Path for the output CSV file
output_csv_path = '/content/output_imgv1/extract_text.csv'

# Process the file and write to a CSV
with open(input_file_path, 'r') as infile, open(output_csv_path, 'w', newline='', encoding='utf-8') as outfile:
    csv_writer = csv.writer(outfile)
    # Writing headers to the CSV file
    csv_writer.writerow(['Filename', 'Transcriptions'])

    for line in infile:
        file_name, transcriptions = process_line(line)
        csv_writer.writerow([file_name, transcriptions])

print("CSV file created successfully.")
```
